```python
import os

import numpy as np
import spacy

from spacy import displacy

from scripts.jsonl import read_jsonl
```

```python
MODEL_PATH = "training/spancat/model-best"

DATA_PATH = "../../data/"
TEST_DATA = os.path.join(DATA_PATH, "jsonl/test.jsonl")
```

```python
test_data = list(read_jsonl(TEST_DATA))
```

```python
nlp = spacy.load(MODEL_PATH)
```

```python
def spancat_visualize(nlp: spacy.language.Language, text: str):
    """Visualize the named entities in the text using the spancat model"""
    doc = nlp(text)

    spans = []
    for span in doc.spans["sc"]:
        spans.append(
            {"start_token": span.start, "end_token": span.end, "label": span.label_}
        )

    data = {
        "text": text,
        "spans": spans,
        "tokens": [token.text for token in doc],
    }

    displacy.render(data, style="span", manual=True)
```

```python
example = np.random.choice(test_data)
spancat_visualize(nlp, example["sentences"])
```
